<span style='font-size:30px;'>Importing Libraries</span>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

<span style='font-size:30px;'>Reading CSV</span>

```python
data = pd.read_csv('std_marks_ds_II.csv')

data['reg_no'] = data['reg_no'].astype('int')

del data['s_no']

data = data.rename(columns={'NME':'Non_Maj_Elec'})

data
```

<span style='font-size:20px;'>Students who are not Detained:</span>

```python
df = data
df
```

```python
rows_to_drop = df[['lang','eng','stats_1','rdbms','pl_sql_lab','nme']].apply(lambda row: 'ab' in row.values, axis=1)
df = df[~rows_to_drop].reset_index(drop= True)
df
```

```python
df[['lang','eng','stats_1','rdbms','pl_sql_lab','nme']] = df[['lang','eng','stats_1','rdbms','pl_sql_lab','nme']].astype('int')
df = df.reset_index(names= 's_no')
df['s_no'] = df['s_no']+1
df
```

<span style='font-size:30px;'>Data Transformation</span>

<span style='font-size:20px;'>Adding "Total Mark" Column</span>

```python
df['total_mark'] = df['lang']+df['eng']+df['stats_1']+df['rdbms']+df['pl_sql_lab']+df['nme']
```

<span style='font-size:20px;'>Adding "Percentage" Column</span>

```python
df['percent'] = round(df['total_mark']/600*100).astype('str').str.split('.').str[0].astype('int')
```

<span style='font-size:20px;'>Assigning Grades</span>

```python
def assigngrade(percentage):
    if percentage == 100:
        return 'O'
    elif 90 <= percentage < 100:
        return 'O'
    elif 80 <= percentage < 90:
        return 'D+'
    elif 75 <= percentage < 80:
        return 'D'
    elif 70 <= percentage < 75:
        return 'A+'
    elif 60 <= percentage < 70:
        return 'A'
    else:
        return 'B'
    
df['grade'] = df['percent'].apply(assigngrade)
```

<span style='font-size:20px;'>Assigning Result</span>

Checks if the Student has got pass in all the Subjects if yes then it marked as 'P' else 'F'

```python
# Function to assign result
def assignresult(row):
    if (
        row['lang'] >= 40 and
        row['eng'] >= 40 and
        row['stats_1'] >= 40 and
        row['rdbms'] >= 40 and
        row['pl_sql_lab'] >= 40 and
        row['nme'] >= 40
    ):
        return 'P'  # Pass
    else:
        return 'F'  # Fail

# Apply the function to each row
df['result'] = df.apply(assignresult, axis=1)
```

<span style='font-size:20px;'>Top 5 Performers</span>

```python
a = df.sort_values(by='total_mark', ascending=False).head(5).reset_index(drop= True)
del a['s_no']
a
```

```python
a = a.reset_index(names= 's_no')
a['s_no'] = a['s_no']+1
a
```

<span style='font-size:20px;'>Number of Students Per Grade</span>

```python
grs = df.groupby('grade')['std_name'].size().reset_index(name= 'total')
grs = grs.reset_index(names= 's_no')
grs['s_no'] = grs['s_no']+1
grs
```

<span style='font-size:30px;'>Finding Average Mark, Maximum Mark and Pass Percent for Subjects</span>

<span style='font-size:20px;'>Assigning Results for Each Students for different Subjects</span>

```python
b = df.rename(columns={'lang':'Language','eng':'English','stats_1':'Statistics I','rdbms':'RDBMS','pl_sql_lab':'PL/SQL-Lab','nme':'NME'})
c = b.melt(id_vars=['s_no'], value_vars=['Language','English','Statistics I','RDBMS','PL/SQL-Lab','NME'],
        var_name='Subjects', value_name='Score')
c
```

```python
def assignresult_2(result):
    if result > 39:
        return "P" 
    else:
        return "F"
    
c['result'] = c['Score'].apply(assignresult_2)
c
```

<span style='font-size:20px;'>Assigning Pass Percentage for each Subject</span>

```python
pass_percent = (
    c.groupby('Subjects')
    .apply(lambda x: round((x['result'] == 'P').sum() / len(x) * 100))
    .reset_index(name='pass_percent')
)

pass_percent
```

<span style='font-size:20px;'>Finding Average Mark, Maximum Mark</span>

```python
avg_max_marks = c.groupby('Subjects')['Score'].agg(['mean','max']).reset_index()
avg_max_marks['mean'] = (avg_max_marks['mean'].astype('str').str.split('.').str[0]).astype('int')
avg_max_marks
```

```python
avg_max_marks = pd.merge(avg_max_marks, pass_percent, on='Subjects')
avg_max_marks
```

```python
avg_max_marks['S_id'] = range(1, len(avg_max_marks) + 1)
avg_max_marks['S_id'] = avg_max_marks['S_id']+100
```

```python
avg_max_marks = avg_max_marks[['S_id','Subjects','mean','max','pass_percent']]
avg_max_marks
```

<span style='font-size:20px;'>Number of Students who Passed All Subjects and it's Percentage</span>
